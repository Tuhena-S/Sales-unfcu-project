```python
import pandas as pd
import os

working_directory = os.getcwd()
print(working_directory)
```

```python
# Load datasets
path_1 = working_directory + '/Unfcu_Data/Roster.csv'
roster_df = pd.read_csv(path_1)
roster_df.head()

path_2 = working_directory + '/Unfcu_Data/Sales_eligibility.csv'
eligibility_df = pd.read_csv(path_2)
eligibility_df.head()

path_3 = working_directory + '/Unfcu_Data/sales_transactions.csv'
sales_df = pd.read_csv(path_3)
sales_df
```

```python
duplicates = sales_df[sales_df.duplicated()]

# Display duplicates for review
print(f"Number of duplicate rows: {len(duplicates)}")
if not duplicates.empty:
    print("Duplicate rows:")
    print(duplicates)

# Remove duplicate rows
sales_df = sales_df.drop_duplicates()

sales_df

```

```python
# Convert dates to datetime objects for comparison
roster_df['HIRE DATE'] = pd.to_datetime(roster_df['HIRE DATE'])
sales_df['SALES DATE'] = pd.to_datetime(sales_df['SALES DATE'])

roster_df
sales_df
```

```python
# Merge roster with eligibility
roster_eligibility_df = pd.merge(roster_df, eligibility_df, on="TITLE", how="left")

roster_eligibility_df
```

```python
# Merge with sales data
merged_df = pd.merge(sales_df, roster_eligibility_df, on="NAME", how="left")

merged_df
```

```python
# Filter rows based on eligibility
eligible_data = merged_df[
    ((merged_df["PRODUCT"] == "RESEARCH") & (merged_df["RESEARCH"] == "Yes")) |
    ((merged_df["PRODUCT"] == "TICKET") & (merged_df["TICKET"] == "Yes")) |
    ((merged_df["PRODUCT"] == "MEDIA") & (merged_df["MEDIA"] == "Yes")) |
    ((merged_df["PRODUCT"] == "CONSULTING") & (merged_df["CONSULTING"] == "Yes")) |
    ((merged_df["PRODUCT"] == "FORECASTS") & (merged_df["FORECASTS"] == "Yes"))
]
eligible_data
```

```python

# Further filter based on hire date
final_cleaned_df = eligible_data[eligible_data['SALES DATE'] >= eligible_data['HIRE DATE']]

final_cleaned_df
```

```python
# Save cleaned data

final_cleaned_df.to_csv("cleaned_sales_data_with_hire_date.csv", index=False)

# Display first few rows of the cleaned data
print(final_cleaned_df)

```
